```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
```

# Safe mode

```python
fs, x = scipy.io.wavfile.read('/home/daniel/Descargas/SDRSharp_20200730_202519Z_8414874620Hz_IQ_2M5_dec8.wav',
                             mmap = True)
```

```python
N = 2**14
L = int(20 * fs)//N*N
f = np.average(np.abs(np.fft.fft(x[:L,0].reshape((-1,N)) + 1j * x[:L,1].reshape((-1,N))))**2, axis = 0)
```

```python
freq = np.fft.fftshift(np.fft.fftfreq(N, 1/fs))
plt.figure(figsize = (14,8), facecolor = 'w')
plt.plot(freq * 1e-3, 10 * np.log10(np.fft.fftshift(f)) - 81)
plt.grid()
plt.title('Mars 2020 X-band safe mode telemetry spectrum')
plt.xlabel('Frequency (kHz)')
plt.ylabel('PSD (dB)');
```

```python
sel = (freq >= -20e3) & (freq <= 40e3)
plt.figure(figsize = (14,8), facecolor = 'w')
plt.plot(freq[sel] * 1e-3, 10 * np.log10(np.fft.fftshift(f))[sel] - 81)
plt.grid()
plt.title('Mars 2020 X-band safe mode telemetry spectrum')
plt.xlabel('Frequency (kHz)')
plt.ylabel('PSD (dB)');
```
