```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
```

```python
x = np.fromfile('bits_1kbaud.f32', dtype = 'float32')[500:]
x_diff = -x[1:]*x[:-1]
```

```python
corr = scipy.signal.correlate(x, x, mode = 'full')
plt.plot(corr[corr.size//2:][:100])
```

```python
corr_diff = scipy.signal.correlate(x_diff, x_diff, mode = 'full')
plt.plot(corr_diff[corr_diff.size//2:][:100])
```

```python
np.average(x_diff)
```
